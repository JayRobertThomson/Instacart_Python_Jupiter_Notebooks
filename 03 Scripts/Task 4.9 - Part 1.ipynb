```python
# Check for mixed data type columns

for col in cust_df.columns:
    weird = (cust_df[[col]].applymap(type) != cust_df[[col]].iloc[0].apply(type)).any(axis=1)
    if len(cust_df[weird]) > 0:
        print(col)
```
