# Combining & Exporting Data
## Contents: Concatenate df, Mergeing df, Appending df, JOINS

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# define data for a dictionary for January 2022

data_1 = {'customer_id': ['1234', '2345', '3456', '4567'],
       'month_id' : ['Jan-22', 'Jan-22', 'Jan-22', 'Jan-22' ],
       'purchased_meat' : [12, 3, 8, 0],
       'Purchased_alcohol' : [7, 9, 11, 27],
       'purchased_snacks' : [1, 9, 3, 22]}
```

```python
# define data for a dictionary for February 2022

data_2 = {'customer_id': ['1234', '2345', '3456', '4567'],
       'month_id' : ['Feb-22', 'Feb-22', 'Feb-22', 'Feb-22' ],
       'purchased_meat' : [1, 4, 6, 3],
       'Purchased_alcohol' : [12, 3, 1, 7],
       'purchased_snacks' : [3, 5, 7, 2]}
```

```python
# Convert the dictionary into DataFrame 

df = pd.DataFrame (data_1, index = [0,1,2,3])
```

```python
# Convert the dictionary into DataFrame 

df_1 = pd.DataFrame (data_2, index = [0,1,2,3])
```

```python
df
```

```python
df_1
```

```python
# Create a list that contains our dataframes

frames = [df,df_1]
```

```python
# Check the output

frames
```

```python
# Check the data types 

type (frames)
```

```python
# Concatenate the dataframes using default options

df_concat = pd.concat(frames)
```

```python
# Check the output

df_concat
```

```python
# Concatenate the dataframes using the axis = 1 --> create a wide format as an output

df_concat = pd.concat (frames, axis = 1)
```

```python
# Check the output

df_concat
```

# Merging Data

```python
# Merge df and df_2 with customer_id as key 

df_merged = df.merge(df_2, on = ['customer_id'])
```

```python
df_merged
```

```python
# Merge df and df_2 with customer_id and month_id as key 

df_merged = df.merge(df_2, on = ['customer_id', 'month_id'])
```

```python
df_merged
```

```python
# Merge df and df_2 using customer_id and month as a keys, add a merge flag

df_merged = df.merge(df_2, on = ['customer_id', 'month_id'], indicator = True)
```

```python
df_merged
```

```python
df_merged['_merge'].value_counts()
```

```python
# Test merge without overwriting

pd.merge(df,df_2, on = ['customer_id', 'month_id'], indicator = True)
```
